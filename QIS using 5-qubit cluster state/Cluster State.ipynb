```python
qc = QuantumCircuit(5,4)



for i in range(1,5):
    qc.h(i)
for i in range(1,4):
    qc.cz(i,i+1)
qc.barrier()

qc.h(1)
qc.h(4)


qc.measure([1,2,3,4],[0,1,2,3])



qc.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend = backend, shots = 1024)

results = job.result()
plot_histogram(results.get_counts(qc))
```
